```python
"""
================================================================================
  FPL TWEET SENTIMENT ANALYSIS - Complete End-to-End ML Pipeline
  Dataset: https://www.kaggle.com/datasets/prasad22/fpl-tweets-dataset
  Author: Final Year Engineering Mini-Project
================================================================================
"""

# ─── IMPORTS ──────────────────────────────────────────────────────────────────
import re
import warnings
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import seaborn as sns
from collections import Counter

# NLP
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer

# Sklearn
from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import MultinomialNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import LinearSVC
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score,
    f1_score, classification_report, confusion_matrix
)
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import LabelEncoder
from scipy.sparse import hstack, csr_matrix

# Word Cloud
from wordcloud import WordCloud

warnings.filterwarnings("ignore")

# ─── NLTK DOWNLOADS ───────────────────────────────────────────────────────────
print("📦 Downloading NLTK resources...")
nltk.download('all')
for pkg in ["punkt", "stopwords", "wordnet", "omw-1.4", "vader_lexicon"]:
    nltk.download(pkg, quiet=True)

from nltk.sentiment.vader import SentimentIntensityAnalyzer

# ══════════════════════════════════════════════════════════════════════════════
# STEP 1: DATA LOADING & UNDERSTANDING
# ══════════════════════════════════════════════════════════════════════════════

def load_data(filepath: str) -> pd.DataFrame:
    """
    Load CSV dataset and display basic info.
    Replace 'filepath' with the path to your downloaded Kaggle CSV.
    """
    print("\n" + "="*70)
    print("STEP 1: DATA LOADING & UNDERSTANDING")
    print("="*70)

    df = pd.read_csv(filepath)

    print(f"\n📐 Dataset Shape: {df.shape}")
    print(f"\n📋 Column Names:\n{df.columns.tolist()}")
    print(f"\n🔢 Data Types:\n{df.dtypes}")
    print(f"\n📊 First 10 Rows:\n{df.head(10)}")
    print(f"\n❓ Missing Values:\n{df.isnull().sum()}")

    return df


def generate_sentiment_labels(df: pd.DataFrame, text_col: str) -> pd.DataFrame:
    """
    If no sentiment column exists, use VADER to auto-label tweets.
    VADER is a rule-based sentiment analyser tuned for social media text.
    """
    print("\n🤖 No sentiment labels found — using VADER to auto-label...")
    sia = SentimentIntensityAnalyzer()

    def vader_label(text: str) -> str:
        score = sia.polarity_scores(str(text))["compound"]
        if score >= 0.05:
            return "Positive"
        elif score <= -0.05:
            return "Negative"
        else:
            return "Neutral"

    df["sentiment"] = df[text_col].apply(vader_label)
    print(f"\n✅ Sentiment Distribution:\n{df['sentiment'].value_counts()}")
    return df


def eda(df: pd.DataFrame, text_col: str, sentiment_col: str) -> None:
    """Exploratory Data Analysis."""
    print("\n" + "="*70)
    print("STEP 1b: EXPLORATORY DATA ANALYSIS")
    print("="*70)

    # Tweet length
    df["tweet_length"] = df[text_col].astype(str).apply(len)
    df["word_count"] = df[text_col].astype(str).apply(lambda x: len(x.split()))

    print(f"\n📏 Tweet Length Stats:\n{df['tweet_length'].describe()}")

    fig, axes = plt.subplots(1, 3, figsize=(18, 5))
    fig.suptitle("FPL Tweet EDA", fontsize=16, fontweight="bold")

    # Sentiment distribution
    sentiment_counts = df[sentiment_col].value_counts()
    colors = ["#2ecc71", "#e74c3c", "#3498db"]
    axes[0].bar(sentiment_counts.index, sentiment_counts.values, color=colors)
    axes[0].set_title("Sentiment Distribution")
    axes[0].set_xlabel("Sentiment")
    axes[0].set_ylabel("Count")
    for i, v in enumerate(sentiment_counts.values):
        axes[0].text(i, v + 10, str(v), ha="center", fontweight="bold")

    # Tweet length distribution
    axes[1].hist(df["tweet_length"], bins=40, color="#9b59b6", edgecolor="white")
    axes[1].set_title("Tweet Length Distribution")
    axes[1].set_xlabel("Character Count")
    axes[1].set_ylabel("Frequency")

    # Word count distribution
    axes[2].hist(df["word_count"], bins=30, color="#e67e22", edgecolor="white")
    axes[2].set_title("Word Count Distribution")
    axes[2].set_xlabel("Word Count")
    axes[2].set_ylabel("Frequency")

    plt.tight_layout()
    plt.savefig("eda_plots.png", dpi=150, bbox_inches="tight")
    plt.show()
    print("✅ EDA plots saved as 'eda_plots.png'")


# ══════════════════════════════════════════════════════════════════════════════
# STEP 2: DATA PREPROCESSING (NLP PIPELINE)
# ══════════════════════════════════════════════════════════════════════════════

STOP_WORDS = set(stopwords.words("english"))
LEMMATIZER = WordNetLemmatizer()


def clean_tweet(text: str) -> str:
    """
    Full NLP cleaning pipeline for a single tweet:
    1. Lowercase
    2. Remove URLs
    3. Remove @mentions
    4. Remove hashtag symbol (keep text)
    5. Remove emojis and non-ASCII
    6. Remove punctuation and digits
    7. Tokenize
    8. Remove stopwords
    9. Lemmatize
    """
    text = str(text).lower()
    text = re.sub(r"http\S+|www\S+", "", text)          # URLs
    text = re.sub(r"@\w+", "", text)                     # @mentions
    text = re.sub(r"#(\w+)", r"\1", text)               # #hashtag → keep word
    text = text.encode("ascii", "ignore").decode("ascii") # emojis / non-ASCII
    text = re.sub(r"[^a-z\s]", "", text)                # punctuation & digits
    text = re.sub(r"\s+", " ", text).strip()            # extra whitespace

    tokens = word_tokenize(text)
    tokens = [LEMMATIZER.lemmatize(t) for t in tokens
              if t not in STOP_WORDS and len(t) > 2]
    return " ".join(tokens)


def preprocess_data(df: pd.DataFrame, text_col: str) -> pd.DataFrame:
    """Apply cleaning pipeline and show before/after examples."""
    print("\n" + "="*70)
    print("STEP 2: DATA PREPROCESSING")
    print("="*70)

    df["cleaned_text"] = df[text_col].apply(clean_tweet)

    print("\n🔍 BEFORE vs AFTER Cleaning (5 examples):")
    print("-" * 70)
    for i in range(min(5, len(df))):
        print(f"BEFORE: {df[text_col].iloc[i]}")
        print(f"AFTER : {df['cleaned_text'].iloc[i]}")
        print("-" * 70)

    return df


# ══════════════════════════════════════════════════════════════════════════════
# STEP 3: FEATURE ENGINEERING
# ══════════════════════════════════════════════════════════════════════════════

def extract_additional_features(df: pd.DataFrame, text_col: str) -> np.ndarray:
    """
    Additional hand-crafted features:
    - tweet_length : raw character count
    - word_count   : number of words
    - hashtag_count: number of hashtags
    - mention_count: number of @mentions
    """
    features = pd.DataFrame()
    features["tweet_length"] = df[text_col].astype(str).apply(len)
    features["word_count"]   = df[text_col].astype(str).apply(lambda x: len(x.split()))
    features["hashtag_count"]= df[text_col].astype(str).apply(lambda x: len(re.findall(r"#\w+", x)))
    features["mention_count"]= df[text_col].astype(str).apply(lambda x: len(re.findall(r"@\w+", x)))
    return features.values


def build_features(df: pd.DataFrame, original_text_col: str):
    """
    Combine TF-IDF (on cleaned text) + additional numeric features.

    WHY TF-IDF over Bag-of-Words?
    ─────────────────────────────
    BoW simply counts word occurrences, giving equal weight to common words
    like 'the', 'is'. TF-IDF down-weights words that appear in MANY documents
    (low discriminative power) and up-weights words that are rare and
    document-specific — much more informative for sentiment classification.
    """
    print("\n" + "="*70)
    print("STEP 3: FEATURE ENGINEERING")
    print("="*70)

    tfidf = TfidfVectorizer(
        max_features=5000,
        ngram_range=(1, 2),     # unigrams + bigrams
        sublinear_tf=True,      # log-scaling of TF
        min_df=2                # ignore very rare terms
    )
    X_tfidf = tfidf.fit_transform(df["cleaned_text"])
    print(f"✅ TF-IDF shape: {X_tfidf.shape}")

    X_extra = extract_additional_features(df, original_text_col)
    X_extra_sparse = csr_matrix(X_extra)

    X_combined = hstack([X_tfidf, X_extra_sparse])
    print(f"✅ Combined feature matrix shape: {X_combined.shape}")

    return X_combined, tfidf


# ══════════════════════════════════════════════════════════════════════════════
# STEP 4: TRAIN-TEST SPLIT
# ══════════════════════════════════════════════════════════════════════════════

def split_data(X, y):
    """Stratified 80/20 split to preserve class distribution."""
    print("\n" + "="*70)
    print("STEP 4: TRAIN-TEST SPLIT (80/20, Stratified)")
    print("="*70)

    le = LabelEncoder()
    y_enc = le.fit_transform(y)

    X_train, X_test, y_train, y_test = train_test_split(
        X, y_enc, test_size=0.2, random_state=42, stratify=y_enc
    )
    print(f"Training samples : {X_train.shape[0]}")
    print(f"Testing samples  : {X_test.shape[0]}")
    return X_train, X_test, y_train, y_test, le


# ══════════════════════════════════════════════════════════════════════════════
# STEP 5: MODEL BUILDING
# ══════════════════════════════════════════════════════════════════════════════

def build_models() -> dict:
    """
    Return a dictionary of untrained classifiers.

    WHY NAIVE BAYES FOR TEXT?
    ─────────────────────────
    MultinomialNB assumes feature independence (word counts independent of each
    other). While naive, this works surprisingly well for text because word
    co-occurrences are sparse and conditional independence is roughly valid.
    It's fast, interpretable, and requires little data to generalise.
    """
    return {
        "Logistic Regression": LogisticRegression(
            max_iter=1000, C=1.0, solver="lbfgs", multi_class="auto", random_state=42
        ),
        "Naive Bayes": MultinomialNB(alpha=0.5),
        "Random Forest": RandomForestClassifier(
            n_estimators=100, random_state=42, n_jobs=-1
        ),
        "SVM (LinearSVC)": LinearSVC(C=1.0, max_iter=2000, random_state=42),
    }


# ══════════════════════════════════════════════════════════════════════════════
# STEP 6: MODEL EVALUATION
# ══════════════════════════════════════════════════════════════════════════════

def evaluate_model(name: str, model, X_train, X_test, y_train, y_test, le, results: list):
    """Train, evaluate, and store results for one model."""
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    acc  = accuracy_score(y_test, y_pred)
    prec = precision_score(y_test, y_pred, average="weighted", zero_division=0)
    rec  = recall_score(y_test, y_pred, average="weighted", zero_division=0)
    f1   = f1_score(y_test, y_pred, average="weighted", zero_division=0)

    results.append({
        "Model": name, "Accuracy": acc,
        "Precision": prec, "Recall": rec, "F1-Score": f1
    })

    print(f"\n{'─'*50}")
    print(f"📊 {name}")
    print(f"{'─'*50}")
    print(f"  Accuracy : {acc:.4f}")
    print(f"  Precision: {prec:.4f}")
    print(f"  Recall   : {rec:.4f}")
    print(f"  F1-Score : {f1:.4f}")
    print(f"\nClassification Report:\n"
          f"{classification_report(y_test, y_pred, target_names=le.classes_, zero_division=0)}")

    # Confusion matrix
    cm = confusion_matrix(y_test, y_pred)
    plt.figure(figsize=(6, 4))
    sns.heatmap(
        cm, annot=True, fmt="d", cmap="Blues",
        xticklabels=le.classes_, yticklabels=le.classes_
    )
    plt.title(f"Confusion Matrix — {name}")
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.tight_layout()
    plt.savefig(f"cm_{name.replace(' ', '_').replace('(', '').replace(')', '')}.png",
                dpi=150, bbox_inches="tight")
    plt.show()

    return model


# ══════════════════════════════════════════════════════════════════════════════
# STEP 7: HYPERPARAMETER TUNING
# ══════════════════════════════════════════════════════════════════════════════

def tune_best_model(X_train, y_train) -> LogisticRegression:
    """
    GridSearchCV on Logistic Regression (commonly best on text tasks).
    Tune: C (regularisation strength) and solver.
    """
    print("\n" + "="*70)
    print("STEP 7: HYPERPARAMETER TUNING (GridSearchCV on Logistic Regression)")
    print("="*70)

    param_grid = {
        "C": [0.01, 0.1, 1, 5, 10],
        "solver": ["lbfgs", "liblinear"],
    }

    grid = GridSearchCV(
        LogisticRegression(max_iter=1000, multi_class="auto", random_state=42),
        param_grid, cv=5, scoring="f1_weighted", n_jobs=-1, verbose=1
    )
    grid.fit(X_train, y_train)

    print(f"\n✅ Best Params  : {grid.best_params_}")
    print(f"✅ Best CV F1   : {grid.best_score_:.4f}")
    return grid.best_estimator_


# ══════════════════════════════════════════════════════════════════════════════
# STEP 8: VISUALIZATIONS
# ══════════════════════════════════════════════════════════════════════════════

def generate_wordcloud(df: pd.DataFrame, sentiment_col: str) -> None:
    """Word clouds per sentiment class."""
    print("\n☁️  Generating Word Clouds...")
    sentiments = df[sentiment_col].unique()
    fig, axes = plt.subplots(1, len(sentiments), figsize=(7 * len(sentiments), 5))
    if len(sentiments) == 1:
        axes = [axes]

    cmap_map = {"Positive": "Greens", "Negative": "Reds", "Neutral": "Blues"}

    for ax, sent in zip(axes, sentiments):
        text = " ".join(df[df[sentiment_col] == sent]["cleaned_text"].dropna().tolist())
        if not text.strip():
            continue
        wc = WordCloud(
            width=600, height=400, background_color="white",
            colormap=cmap_map.get(sent, "viridis"), max_words=100
        ).generate(text)
        ax.imshow(wc, interpolation="bilinear")
        ax.axis("off")
        ax.set_title(f"{sent} Tweets", fontsize=14, fontweight="bold")

    plt.tight_layout()
    plt.savefig("wordcloud.png", dpi=150, bbox_inches="tight")
    plt.show()
    print("✅ Word cloud saved as 'wordcloud.png'")


def plot_model_comparison(results: list) -> None:
    """Bar chart comparing all model metrics."""
    df_res = pd.DataFrame(results)
    df_melt = df_res.melt(id_vars="Model", var_name="Metric", value_name="Score")

    plt.figure(figsize=(12, 6))
    ax = sns.barplot(data=df_melt, x="Model", y="Score", hue="Metric", palette="Set2")
    ax.set_ylim(0, 1.0)
    ax.set_title("Model Performance Comparison", fontsize=15, fontweight="bold")
    ax.set_xlabel("")
    ax.set_ylabel("Score")
    plt.xticks(rotation=15)
    plt.legend(loc="lower right")
    plt.tight_layout()
    plt.savefig("model_comparison.png", dpi=150, bbox_inches="tight")
    plt.show()
    print("✅ Model comparison chart saved.")


# ══════════════════════════════════════════════════════════════════════════════
# STEP 9: BEST MODEL SELECTION
# ══════════════════════════════════════════════════════════════════════════════

def select_best_model(results: list, trained_models: dict) -> tuple:
    """Select model with highest weighted F1-score."""
    print("\n" + "="*70)
    print("STEP 9: FINAL MODEL SELECTION")
    print("="*70)

    df_res = pd.DataFrame(results).sort_values("F1-Score", ascending=False)
    print(f"\n📊 Model Rankings:\n{df_res.to_string(index=False)}")

    best_name = df_res.iloc[0]["Model"]
    print(f"\n🏆 Best Model: {best_name} (F1={df_res.iloc[0]['F1-Score']:.4f})")
    print("""
Why this model performed best:
  • SVM / Logistic Regression excel on high-dimensional sparse TF-IDF features
    because they find maximum-margin hyperplanes in the feature space.
  • Logistic Regression with L2 regularisation avoids overfitting on noisy tweets.
  • Both are efficient and well-suited for multi-class text classification.
    """)
    return best_name, trained_models[best_name]


# ══════════════════════════════════════════════════════════════════════════════
# STEP 10: PREDICTION SYSTEM
# ══════════════════════════════════════════════════════════════════════════════

class FPLSentimentPredictor:
    """
    Encapsulates the full prediction pipeline:
      raw tweet → clean → TF-IDF + extra features → best model → sentiment label
    """

    def __init__(self, model, tfidf: TfidfVectorizer, le: LabelEncoder,
                 original_text_col: str = "tweet"):
        self.model = model
        self.tfidf = tfidf
        self.le = le
        self.original_text_col = original_text_col

    def predict_tweet(self, text: str) -> str:
        cleaned = clean_tweet(text)
        X_tfidf = self.tfidf.transform([cleaned])
        extra = extract_additional_features(
            pd.DataFrame({self.original_text_col: [text]}),
            self.original_text_col
        )
        X = hstack([X_tfidf, csr_matrix(extra)])
        pred = self.model.predict(X)[0]
        label = self.le.inverse_transform([pred])[0]
        return label

    def predict_batch(self, texts: list) -> list:
        return [self.predict_tweet(t) for t in texts]


def demo_predictions(predictor: FPLSentimentPredictor) -> None:
    """Test prediction system on sample FPL tweets."""
    print("\n" + "="*70)
    print("STEP 10: SAMPLE PREDICTIONS")
    print("="*70)

    sample_tweets = [
        "Salah is on fire this season! Triple captain was the best decision ever! 🔥",
        "My entire team is injured. This is an absolute disaster week for FPL.",
        "Gameweek 15 deadline is tomorrow. Still unsure about my captain pick.",
        "Haaland with another hat-trick! FPL managers are going crazy!",
        "Another blank gameweek ruined my rank. I'm done with this game.",
        "Just a normal week, nothing special happened in FPL this gameweek.",
        "Rashford's form is shocking. Sold him at the wrong time as always.",
        "Bought Palmer and he blanked. Classic FPL. At least my bench saved me.",
    ]

    emoji_map = {"Positive": "✅", "Negative": "❌", "Neutral": "⚪"}
    print(f"\n{'Tweet':<60} {'Sentiment':<12}")
    print("─" * 74)
    for tweet in sample_tweets:
        label = predictor.predict_tweet(tweet)
        print(f"{tweet[:58]:<60} {emoji_map.get(label, '?')} {label}")


# ══════════════════════════════════════════════════════════════════════════════
# BONUS: TWEET CATEGORY CLASSIFIER (Injury / Transfer / Match / Opinion)
# ══════════════════════════════════════════════════════════════════════════════

CATEGORY_KEYWORDS = {
    "Injury News":   ["injured", "injury", "doubt", "ruled out", "fit", "hamstring",
                      "muscle", "knock", "return", "unavailable"],
    "Transfers":     ["transfer", "sign", "sold", "bought", "deal", "fee", "move",
                      "loan", "contract", "joins"],
    "Match Updates": ["goal", "assist", "hat-trick", "score", "win", "lose", "draw",
                      "clean sheet", "red card", "penalty", "gameweek", "blank"],
    "Opinion":       ["think", "should", "captain", "chip", "bench", "strategy",
                      "rank", "template", "differential", "wildcard"],
}


def classify_category(text: str) -> str:
    """Rule-based category classifier using keyword matching."""
    text_lower = text.lower()
    for category, keywords in CATEGORY_KEYWORDS.items():
        if any(kw in text_lower for kw in keywords):
            return category
    return "General"


# ══════════════════════════════════════════════════════════════════════════════
# MAIN PIPELINE
# ══════════════════════════════════════════════════════════════════════════════

def main(filepath: str):
    """Full end-to-end pipeline."""

    # ── Step 1: Load & Understand ──────────────────────────────────────────
    df = load_data(filepath)

    # Detect text column (common names in the Kaggle dataset)
    text_col = None
    for col in ["tweet", "text", "Tweet", "Text", "content"]:
        if col in df.columns:
            text_col = col
            break
    if text_col is None:
        text_col = df.columns[0]
        print(f"⚠️  Auto-detected text column: '{text_col}'")

    # Detect or create sentiment column
    sentiment_col = None
    for col in ["sentiment", "Sentiment", "label", "Label"]:
        if col in df.columns:
            sentiment_col = col
            break
    if sentiment_col is None:
        df = generate_sentiment_labels(df, text_col)
        sentiment_col = "sentiment"

    # Drop rows where text is missing
    df.dropna(subset=[text_col], inplace=True)
    df.reset_index(drop=True, inplace=True)

    # ── Step 1b: EDA ──────────────────────────────────────────────────────
    eda(df, text_col, sentiment_col)

    # ── Step 2: Preprocess ────────────────────────────────────────────────
    df = preprocess_data(df, text_col)

    # ── Step 3: Features ──────────────────────────────────────────────────
    X, tfidf = build_features(df, text_col)
    y = df[sentiment_col]

    # ── Step 4: Split ─────────────────────────────────────────────────────
    X_train, X_test, y_train, y_test, le = split_data(X, y)

    # ── Step 5 & 6: Train & Evaluate all models ───────────────────────────
    print("\n" + "="*70)
    print("STEPS 5 & 6: MODEL TRAINING & EVALUATION")
    print("="*70)

    models = build_models()
    results = []
    trained_models = {}

    # Note: MultinomialNB requires non-negative features; use TF-IDF only for NB
    X_tfidf_only = tfidf.transform(df["cleaned_text"])
    X_tr_nb, X_te_nb, _, _ = train_test_split(
        X_tfidf_only, y, test_size=0.2, random_state=42, stratify=y
    )
    _, _, y_tr_enc, y_te_enc = train_test_split(
        X, le.transform(y), test_size=0.2, random_state=42, stratify=le.transform(y)
    )

    for name, model in models.items():
        if name == "Naive Bayes":
            trained = evaluate_model(name, model, X_tr_nb, X_te_nb, y_tr_enc, y_te_enc, le, results)
        else:
            trained = evaluate_model(name, model, X_train, X_test, y_train, y_test, le, results)
        trained_models[name] = trained

    # ── Step 7: Hyperparameter Tuning ─────────────────────────────────────
    tuned_lr = tune_best_model(X_train, y_train)
    trained_models["Logistic Regression"] = tuned_lr  # update with tuned version

    # ── Step 8: Visualisations ────────────────────────────────────────────
    print("\n" + "="*70)
    print("STEP 8: VISUALIZATIONS")
    print("="*70)
    generate_wordcloud(df, sentiment_col)
    plot_model_comparison(results)

    # Most common words overall
    all_words = " ".join(df["cleaned_text"].tolist()).split()
    top_words = Counter(all_words).most_common(20)
    words, counts = zip(*top_words)
    plt.figure(figsize=(12, 5))
    plt.barh(list(words)[::-1], list(counts)[::-1], color="#3498db")
    plt.title("Top 20 Most Frequent Words", fontweight="bold")
    plt.xlabel("Frequency")
    plt.tight_layout()
    plt.savefig("top_words.png", dpi=150, bbox_inches="tight")
    plt.show()

    # ── Step 9: Best Model Selection ──────────────────────────────────────
    best_name, best_model = select_best_model(results, trained_models)

    # ── Step 10: Prediction System ────────────────────────────────────────
    predictor = FPLSentimentPredictor(best_model, tfidf, le, text_col)
    demo_predictions(predictor)

    # Bonus: Category Classification
    print("\n" + "="*70)
    print("BONUS: TWEET CATEGORY CLASSIFICATION")
    print("="*70)
    sample = [
        "Salah ruled out with hamstring injury for 3 weeks.",
        "City sign striker for £80m from Bundesliga.",
        "Haaland scores a hat-trick! Unbelievable performance tonight!",
        "Should I use my wildcard this gameweek or wait?",
    ]
    for t in sample:
        cat = classify_category(t)
        sent = predictor.predict_tweet(t)
        print(f"  [{cat:<14}] [{sent:<8}] {t}")

    print("\n" + "="*70)
    print("✅ PIPELINE COMPLETE")
    print("="*70)
    print("""
KEY INSIGHTS & EXPLANATIONS
════════════════════════════

1. WHY TF-IDF > BAG OF WORDS?
   BoW counts raw occurrences. TF-IDF additionally penalises words that appear
   in many documents (like 'the', 'is'), making rare, informative words more
   impactful. This results in better class separability for sentiment tasks.

2. WHY NAIVE BAYES FOR TEXT?
   Despite its 'naive' conditional independence assumption, MultinomialNB
   performs excellently because:
   • Text features ARE roughly independent (knowing 'great' is present
     doesn't tell us much about whether 'awesome' is present).
   • It requires very little training data to generalise.
   • It's extremely fast and interpretable.

3. LIMITATIONS:
   • VADER labels are imperfect — a real labelled FPL dataset would be better.
   • Sarcasm ('great, another blank gameweek') is hard to detect.
   • FPL-specific jargon (wildcard, DGW, FH) may not appear in generic corpora.
   • Context window is missing — individual tweets lack surrounding discourse.

4. FUTURE IMPROVEMENTS:
   • Real-time streaming with Twitter/X API (Tweepy or snscrape).
   • BERT fine-tuning for context-aware embeddings.
   • Aspect-based sentiment (player-level sentiment analysis).
   • Dashboard integration (Streamlit / Flask web app).
   • Engagement prediction using tweet metadata (likes, retweets).
    """)

    return predictor


# ══════════════════════════════════════════════════════════════════════════════
# ENTRY POINT
# ══════════════════════════════════════════════════════════════════════════════

if __name__ == "__main__":
    # ─── CHANGE THIS PATH to your downloaded Kaggle CSV ───────────────────
    DATASET_PATH = "FPL_tweets.csv"   # e.g. "data/fpl-tweets-dataset.csv"
    # ──────────────────────────────────────────────────────────────────────

    predictor = main("/FPL_tweets.csv")

    # Interactive single-tweet prediction
    print("\n💬 Enter a tweet to predict sentiment (or 'quit' to exit):")
    while True:
        user_input = input(">>> ")
        if user_input.lower() in ("quit", "exit", "q"):
            break
        sentiment = predictor.predict_tweet(user_input)
        category  = classify_category(user_input)
        print(f"    Sentiment: {sentiment}  |  Category: {category}")
```
